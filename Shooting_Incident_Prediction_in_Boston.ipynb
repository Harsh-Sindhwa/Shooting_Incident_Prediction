# PREDICTION OF SHOOTING INCIDENTS IN BOSTON IN THE YEAR 2019

```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Extraction

```python
#Import the dataset
dataset = pd.read_csv('https://data.boston.gov/dataset/6220d948-eae2-4e4b-8723-2dc8e67722a3/resource/12cb3883-56f5-47de-afa5-3b1cf61b257b/download/tmplczp2c4s.csv')
dataset.head(10) #shows first 10 records
```

#The dataset is publically available and can be dowloaded easily. I opened the file using pandas library by directly using the link and saved it in the form of dataframe in the variable dataset.

# Data Cleanup

```python
#Taking care of missing values
dataset.replace(' ','NaN', inplace=True) #Replace blank (missing values) spaces by Nan in dataset
X = dataset.loc[:,['REPORTING_AREA']]

#Using sklearn library for missing values
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN',strategy='most_frequent',axis=0)
imputer =imputer.fit(X)
dataset.loc[:,['REPORTING_AREA']]=imputer.transform(X)
dataset.head(10)  
```

#There were missing values in the dataset which were filled by the mode (i.e most occured value)in the dataset. I tried not to drop any data so that it doesn't affect the accuracy of model. 

# Data Visualization

```python
# Creating tables according to different years
def dataset_by_year(table,year): 
    year=table['YEAR']== int(year)
    return table.loc[year]
```

```python
#Number of incidents that occured in a year
plt.figure(figsize=(8,6))
for i in [2015,2016,2017,2018,2019]:
    dataset_year=dataset_by_year(dataset,i)
    incident_year=dataset_year.shape
    incidents_year_r=incident_year[0]
    print("Total incidents in " + str(i) + " is " + str(incidents_year_r))
    plt.bar(i,incidents_year_r,label="YEAR " + str(i))
    plt.xlabel("YEAR")
    plt.ylabel("NUMBER OF INCIDENTS")
plt.legend(loc='best')
plt.show()
```

#Using the above results we see that the crime rated have been increasing consistently. The 2019 year has records only upto the 9th month of the year. I have still plotteed it and the crime incidents are higher than the 2015 year still.

```python
#Number of shooting incidents in each year
plt.figure(figsize=(8,6))
shoot_count=0
year_shootings = {}
for i in [2015,2016,2017,2018]:
    dataset_year=dataset_by_year(dataset,i)
    count = (dataset_year['SHOOTING']=='Y').sum()
    year_shootings[i] = count
    print("Total shooting incidents in " + str(i) + " is " + str(count))
    #plt.bar(i,count,label="YEAR " + str(i))
    plt.xlabel("YEAR")
    plt.ylabel("NUMBER OF SHOOTING INCIDENTS")
#plt.legend(loc='best')
plt.plot(year_shootings.keys(), year_shootings.values())
plt.show()
year_shootings
```

#The shooting incidents involved in the years have increased consistently from 2015 to 2017 and there is a bit drop in the year 2018

# Predicting Analytics

```python
X=dataset.iloc[:,2].values
y = dataset.iloc[:,10].values
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X= labelencoder_X.fit_transform(X)
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
X = X[:, np.newaxis]
y = y[:, np.newaxis]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

#Predicting the Test set results
y_pred = regressor.predict(X_test)
```
